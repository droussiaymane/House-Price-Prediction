```python
new_data = new_data.dropna(subset=['prix'])
new_data = new_data.dropna(subset=['nombre de piece'])
new_data = new_data.dropna(subset=['nombre de chambre'])
new_data = new_data.dropna(subset=['etat'])
new_data.head()
```

```python
new_data['prix'] = new_data['prix'].str.replace(r'\D', '', regex=True).astype(float)
new_data.head()
```

```python
new_data['zone'] = new_data['zone'].str.split(' à').str.get(0)
new_data.head()
```

```python
new_data['surface'] = new_data['surface'].str.extract(r'(\d+)').astype(float)
new_data.head()
```

```python
new_data['nombre de piece'] = new_data['nombre de piece'].str.extract(r'(\d+)')
```

```python
new_data['nombre de piece'] = new_data['nombre de piece'].astype(int)
new_data.head()
```

```python
new_data['nombre de chambre'] = new_data['nombre de chambre'].str.extract(r'(\d+)')
new_data['nombre de chambre'] = new_data['nombre de chambre'].astype(int)
new_data.head()
```

```python
new_data.head()
```

```python
allowed_values = ['Bon état', 'Nouveau', 'À rénover']

# Filter the DataFrame to only include rows with the allowed values in the 'etat' column
new_data = new_data[new_data['etat'].isin(allowed_values)]
```

```python
new_data['etat'].unique()
```

```python
# Create dummy variables for the 'etat' column
etat_dummies = pd.get_dummies(new_data['etat'], prefix='etat')

# Drop the original 'etat' column
new_data = new_data.drop('etat', axis=1)

# Concatenate the dummy variables to the DataFrame
new_data = pd.concat([new_data, etat_dummies], axis=1)

# Check the updated DataFrame
print(new_data.head())
```

```python
data_train=new_data
# Calculate the average price per square meter for each zone
data_train['prix par m2'] = data_train['prix'] / data_train['surface']
avg_price_per_sqm_by_zone = data_train.groupby('zone')['prix par m2'].mean()

# Normalize the average price per square meter to obtain coefficients between 0 and 10
min_avg_price = avg_price_per_sqm_by_zone.min()
max_avg_price = avg_price_per_sqm_by_zone.max()
avg_price_range = max_avg_price - min_avg_price
avg_price_per_sqm_by_zone_normalized = (avg_price_per_sqm_by_zone - min_avg_price) / avg_price_range * 10

# Create a new column 'zone_coefficient' based on the 'zone' column
data_train['zone_coefficient'] = data_train['zone'].map(avg_price_per_sqm_by_zone_normalized)

# Calculate the correlation matrix
correlation_matrix = data_train.corr()

data_train.head()
```

```python
zones = data_train["zone"].unique()
zone_coefficients = {}

for zone in zones:
    zone_rows = data_train[data_train["zone"] == zone]
    zone_coefficient = zone_rows["zone_coefficient"].iloc[0]
    zone_coefficients[zone] = zone_coefficient

print(zone_coefficients)
```

```python
# Select only the numerical columns, excluding the one-hot encoded 'etat' columns
numerical_columns = data_train.select_dtypes(include=['number']).columns
numerical_columns = numerical_columns.difference(['etat_Bon état', 'etat_Nouveau', 'etat_À rénover'])

# Function to remove outliers using the IQR method
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data

# Remove outliers for each numerical column
for column in numerical_columns:
    data_train = remove_outliers_iqr(data_train, column)

# Check the updated DataFrame
print(data_train.head())
```
